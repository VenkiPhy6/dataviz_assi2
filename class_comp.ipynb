```r
library(readxl)
library(tidyverse)
```

```r
hpi <- read_excel("./hpi.xlsx", sheet = "1. All countries", range = "A9:L158")
```

```r
prop_names <- hpi |>
    colnames() |>
    tolower() |>
    trimws() |>
    str_replace_all("\\s", "_") |>
    str_replace_all("\\(.*\\)", "") |>
    str_replace_all("_+$", "")
colnames(hpi) <- prop_names
```
